En mecánica cuántica, las partículas como los electrones se describen mediante una función de onda, la cual es una descripción matemática del estado cuántico de un sistema; la naturaleza probabilística de la función de onda da lugar a varios efectos cuánticos. Mientras exista una relación de fase definida entre diferentes estados, se dice que el sistema es coherente. La coherencia es una propiedad fundamental de la mecánica cuántica, y es necesaria para el funcionamiento de las computadoras cuánticas. Sin embargo, cuando un sistema cuántico no está perfectamente aislado, sino en contacto con su entorno, la coherencia decae con el tiempo, un proceso llamado decoherencia cuántica. Como resultado de este proceso, se pierde el comportamiento cuántico relevante. La decoherencia fue introducida por primera vez en 1970 por el físico alemán H. Dieter Zeh  y ha sido objeto de investigación activa desde la década de 1980.

Para analizar un sistema se usa la matriz de densidad, la cual puede describir un sistema cuántico en un estado mixto, lo cual es un conjunto estadístico de varios estados cuánticos. Esto debe contrastarse con un vector de estado único que describe un sistema cuántico en estado puro.

El efecto de la decoherencia en las matrices de densidad es esencialmente la desintegración o desaparición rápida de los elementos fuera de diagonal de la traza parcial de la matriz de densidad del sistema de unión, es decir, la traza con respecto a cualquier base ambiental de la matriz de densidad del sistema combinado y su entorno. La decoherencia convierte irreversiblemente la matriz de densidad "promediada"  desde un estado puro a una mezcla reducida; es esto lo que da la apariencia de un colapso de la función de onda. La ventaja de tomar la traza parcial es que este procedimiento es indiferente a la base ambiental elegida.

La utilidad de la decoherencia radica en su aplicación al análisis de probabilidades, antes y después de la interacción ambiental, y en particular a la desaparición de los términos de interferencia cuántica después de que se ha producido la decoherencia. Si nos preguntamos cuál es la probabilidad de observar el sistema haciendo una __transición__ de $|\psi \rangle$ a 
$|\phi \rangle$ antes de que $| \psi \rangle$ haya interactuado con su entorno, entonces la aplicación de la regla de probabilidad de Born establece que la probabilidad de transición es el módulo al cuadrado del producto escalar de los dos estados.

Para entender la razón por la cual ocurre la decoherencia, tomemos el estado del sistema         $|\phi \rangle$  y el estado del ambiente
$|\epsilon \rangle$

Inicialmente, la matriz de densidad del sistema combinado se puede denotar como:

$$\rho= |antes \rangle  \langle antes|=|\phi \rangle  \langle \phi| \otimes  |\epsilon \rangle  \langle \epsilon| $$

  
Entonces, si la __transición__ ocurre antes de que tenga lugar cualquier interacción entre el sistema y el entorno, el subsistema de entorno no tiene parte y puede rastrearse, dejando la matriz de densidad reducida para el sistema:


$$\rho_{sis}= Tr_{amb}(\rho)=|\phi \rangle  \langle \phi| \langle \epsilon|\epsilon \rangle $$

Ahora la probabilidad de transición antes de la interacción con el entorno se dará como:

$$prob_{antes}(\psi \rightarrow \phi)= \langle \phi| \rho_{sis}| \phi \rangle= \langle \phi|\psi \rangle \langle \psi|\phi \rangle =|\langle \psi|\phi \rangle|^{2}= \sum\limits_{i=1} |\psi^{*}_{i} \phi_{i}|^{2}+\sum\limits_{i,j, i \neq j} \psi^{*}_{i} \psi_{j} \phi^{*}_{j} \phi_{i} $$
 
con $\psi_{i}= \langle i  |   \psi \rangle$ , $\psi^{*}_{i}=\langle \psi| i \rangle$, analogamente para $\phi$

Ahora el caso cuando la transición tiene lugar después de la interacción del sistema con el entorno. La matriz de densidad combinada será:

 ![matriz](http://localhost:8888/view/Desktop/slide_4.jpg)
 
 

$$\rho= |después \rangle  \langle después|= \sum\limits_{i,j} \psi_{i} \psi^{*}_{j} \otimes| i,\epsilon_{i} \rangle \langle j, \epsilon_{j}|=  \sum\limits_{i,j} \psi_{i} \psi^{*}_{j} | i \rangle \langle j| \otimes |\epsilon_{i} \rangle \langle \epsilon_{j}| $$

Para obtener la matriz de densidad reducida del sistema, trazamos sobre el entorno, vemos que los términos fuera de diagonal se desvanecen.

$$\rho_{sis}= Tr_{amb}(\sum\limits_{i,j} \psi_{i} \psi^{*}_{j} | i \rangle \langle j| \otimes |\epsilon_{i} \rangle \langle \epsilon_{j}|)= \sum\limits_{i,j} \psi_{i} \psi^{*}_{j} | i \rangle \langle j| \otimes\langle \epsilon_{i}|\epsilon_{j} \rangle=  \sum\limits_{i,j} \psi_{i} \psi^{*}_{j} | i \rangle \langle j| \delta_{ij}= \sum\limits_{i} |\psi_{i}|^{2} | i \rangle \langle i|$$

De manera similar, la matriz de densidad reducida final después de la transición será $\sum\limits_{j} |\psi_{j}|^{2} | j \rangle \langle j|$
La probabilidad de transición se dará como:
 $$prob_{después}(\psi \rightarrow \phi)= \sum\limits_{i,j} |\psi_{i}|^{2}|\phi_{j}|^{2} \langle j||i \rangle \langle i||j \rangle = \sum\limits_{i}|\psi^{*}_{i}\phi_{i}|^{2}$$
 
Que no tiene ninguna contribución de los términos de interferencia  $\sum\limits_{i,j, i \neq j} \psi^{*}_{i} \psi_{j} \phi^{*}_{j} \phi_{i}$
 	
 En en este programa se analizara el hamiltoniano dado por:
 $$H= \frac{a}{2} \sigma_{z} \otimes  \mathbb{1}_{nxn}+ \mathbb{1}_{2x2}\otimes H_{env, nxn}+ g\times V_{2nx2n}, g	 \in 	\mathbb{R}$$
 
  donde  $H_{env,nxn}$ es el hamiltoniano del ambiente, modelado mediante las matrices GUE (Gaussian unitary ensemble), y 
  $V_{2nx2n}$ es el hamiltoniano de interacción sistema-ambiente usando una matriz GOE (Gaussian orthogonal ensemble), escencialmente son matrices invariantes ante transformaciones unitarias y transformaciones otrogonales respectivamente, cuyos elementos cumplen con cierta distribución gaussiana dada por:
  
  \begin{equation}
P(s)=\begin{cases} 
      \hfill (s \pi /2)\exp (-s^2\pi/4)    \hfill & \text{ortogonal} \\
      \hfill (s^2 32/\pi^2) \exp (-s^2 4/\pi)    \hfill & \text{unitario} \\
  \end{cases}
\end{equation}

Se analiza el hamiltoniano tomando estados aleatorios factorizables
$|\Psi \rangle= |	\uparrow \rangle \otimes |\alpha \rangle$

después se evolucionan como:

$$|\Psi (t) \rangle =  e^{-iHt} |\Psi \rangle$$

De tal manera que la matriz de densidad será:

$$ \rho= |\Psi (t) \rangle \langle \Psi (t)|=e^{-iHt} |\Psi \rangle \langle \Psi| e^{iHt}$$


Luego se calcula la traza parcial respecto a la base del ambiente, para obtener la matriz de densidad del sistema


 $$ \rho_{sis}(t)=Tr_{amb}(\rho(t))$$
 
 Y calcular la pureza como:
 
 $$P(t)=Tra(\rho^{2}_{sis}(t))$$
 
Ya que la matriz de densidad de un estado puro es $\rho=|\psi \rangle\langle\psi|$, entonces para un estado puro P=1, sin embargo los estados mezcla se representan con $\rho_{mezcla}=\sum\limits_{i}p_{i} |\psi_{i} \rangle\langle\psi_{i}|$, 
donde $ \sum\limits_{i}p_{i} =1$ la pureza $ P<1$ 
En nuestro análisis esperamos ver como la pureza de los estados disminuye al enredarse con el ambiente.

El la gráfica anterior se obseva que el parámetro g del Hamiltoniano de interacción influye de manera directamente proporcional
al decaimiento en la pureza del sistema central.

Ahora se analiza el mismo sitema cuando se le agrega un segundo ambiente mas grande.

Sistema original: 
$$H= \frac{a}{2} \sigma_{z} \otimes  \mathbb{1}_{nxn}+ \mathbb{1}_{2x2}\otimes H_{env, nxn}+ g\times V_{2nx2n}$$
$$a=1, g=0.1$$
Sistema nuevo:   
$$H´=\frac{a}{2}\sigma_{z} \otimes  \mathbb{1}_{nxn} \otimes  \mathbb{1}_{nxn}+ \mathbb{1}_{2x2}\otimes H_{env, nxn} \otimes  \mathbb{1}_{nxn} + g \times V_{2nx2n} \otimes \mathbb{1}_{nxn}+ \mathbb{1}_{2x2} \otimes \mathbb{1}_{nxn} \otimes H_{env2, nxn}+G \times \mathbb{1}_{2x2} \otimes V_{n^2 x n^2}$$
$$a=1, g=0.1, G=0.2,$$

En este caso particular se trabajó con  la dimensión de las matrices n=dim=6.


Con $ H_{env2, nxn}$ el hamiltoniano del sistema 2 mas grande, y $V_{n^2 x n^2}$ el hamiltoniano de interacción entre el sitema central y el ambiente 2, ambos modelados mediante matrices GUE.
Lo que se espera observar es que el sistema central se enrede más con el segundo ambiente, observando la pureza del sitema central al obtener la traza parcial respecto al sistema 2.

$$ \rho(t)=e^{-iH't} |\Psi \rangle \langle \Psi| e^{iH't}$$


 $$ \rho_{sis}(t)=Tr_{amb2}(\rho(t))$$
 
 Y calcular la pureza como:
 
 $$P(t)=Tra(\rho^{2}_{sis}(t))$$
  
 Esto se puede entender mediante la __monogamia de enredamiento__:
La monogamia significa que si las partículas A y B están en un estado de enredamiento máximo, por ejemplo, uno de los estados de Bell, entonces la partícula A o B no puede enredarse con la tercera partícula C.

```julia
(t,traza3) = herramienta.H2(1,6,0.1,0.2) #Se asigna el valor a la salida de la lista de tiempo, y 
#la lista con la traza de los estados.
```

La siguiente gráfica muestra el decaimento de la pureza respecto al tiempo, comparando cuando el sistema interectua con un sólo ambiente y cuando interacciona con este mismo ambiente y con otro más grande.

```julia
xlabel("Tiempo");ylabel("Pureza")
title("Comparación del Enredamiento: Sitema-Ambiente_1-Ambiente_2 ")
plot(t,traza2,label="Interacción Sistema-Ambiente_1");
plot(t,traza3,label="Interacción Sistema-Ambiente_1-Ambiente_2");
legend()
#Se observa como el la pureza decae mas rápido cuando se obtiene la traza parcial respecto al ambiente 1.
```

De la gráfica anterior se observa como el sistema central tiende a protejerse debido al acoplamiento entre el ambiente cercano 
y ambiente lejano. También se observa que la pureza debido a la interacción con el sistema cercano decae mas rápido que 
cuando existe la interacción con 2 ambientes.
